#Removing Null Values

```python
df.isnull().sum()
```

#Method 1

```python
df=df.fillna(df.mode().iloc[0])
```

```python
df.isnull().sum()
```

#Method 2

```python
for col in df.select_dtypes(include='O').columns:                            #it is used to remove null values which are in the form of Object
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include=['int64','float64']).columns:              #it is used to remove null values which are in the form of int and float
    df[col]=df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

#Method 3

```python
for col in df.columns:                                          #if we wanted to remove null values permanently we can use inplace=True
    if df[col].dtype=='O':
        df[col].fillna(df[col].mode()[0])
    else:
        df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df.loc[df.Gender.isnull()]                                               #to know how many records has null values in gender column
```

```python
df.loc[df.Gender.isnull(), 'Gender']= 'Male'
```

```python
df.Gender.fillna(value='Male',inplace=True)
```

```python
df.Married.fillna(value="Yes")
```
